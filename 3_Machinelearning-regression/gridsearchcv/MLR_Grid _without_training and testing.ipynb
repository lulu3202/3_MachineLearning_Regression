```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
#converting categorical to numerical value
dataset=pd.get_dummies(dataset,dtype=int, drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
#input variables
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
independent
```

```python
#dependent or output variable
dependent=dataset[["charges"]]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression

# Define the parameter grid
param_grid = {
    'fit_intercept': [True, False],
    'copy_X': [True, False],
    'n_jobs': [-1, None]
}

# Create a GridSearchCV object
grid = GridSearchCV(LinearRegression(), param_grid, refit=True, verbose=3, n_jobs=-1)

# Fitting the model for grid search
grid.fit(independent, dependent)


```

```python

# print best parameter after tuning 
#print(grid.best_params_) 
re=grid.cv_results_
#print(re)
grid_predictions = grid.predict(independent) 
   
# print classification report 
from sklearn.metrics import r2_score
r_score=r2_score(dependent,grid_predictions)

print("The R_score value for best parameter {}:".format(grid.best_params_),r_score)
```
