```python
#Importing library 
import pandas as pd 
```

```python
#Importing dataset - we are calling read csv function from pandas library and we are reading salary data
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#data set is displayed years of exp is input and salary is output
dataset
```

```python
#years of experience (input variable) column is alone extracted and assigned to a variable called independent
independent=dataset[["YearsExperience"]]
```

```python
independent
```

```python
#salary (output variable) column is alone extracted and assigned to a variable called dependent
dependent=dataset[["Salary"]]
dependent
```

```python
#sklearn is a library that supports machine learning; here we are using sklearn library to split training and test data using the below function
#we are assigning variables to the input test data, input training data, output training data and output test data 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0) 
#split input data x into test and train, then output data y into test and train)
```

```python
# these 3 steps comprise the learning phase; from sklearn library we are importing linear regression algorithm - procedure
from sklearn.linear_model import LinearRegression
#regressor is a variabe to which we assign all the functions under linear regression class - procedure
regressor=LinearRegression()
#here we are substituting values where weight and bias is being calulated 
regressor.fit(X_train,y_train)
```

```python
#we are seeing the value of the calulated weight by calling regressor.coef value
weight=regressor.coef_
weight 
```

```python
#we are seeing the value of the calculated bias
bias=regressor.intercept_
bias
```

```python
#cross checking and evaluating the model using y_pred with test data
y_pred=regressor.predict(X_test)
```

```python
#finding r square value and assinging it to r_score and pass actual output and AI's predicted output as parameters
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
#as we get nearly 1, it is a good model, cross checking has worked 
r_score
```

```python
#saving the finalized model into a variable called filename
import pickle
filename="finalized_model_linear.sav"
```

```python
#loading the model into the above filename 
pickle.dump(regressor,open(filename,'wb'))
```

```python
#loading the model to check if it works fine before deploying it 
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
result = loaded_model.predict([[13]])
```
