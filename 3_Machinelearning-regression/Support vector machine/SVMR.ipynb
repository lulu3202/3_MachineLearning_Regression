```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
#changinf state categorical value to numerical value
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
#display columns
dataset.columns
```

```python
#extracting input variables
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
#extracting output variable
dependent=dataset[["Profit"]]
```

```python
#split into training and test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
X_train
```

```python
#improving model performance via standardization
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python

#applying model
from sklearn.svm import SVR
regressor=SVR(C=3000,kernel="linear")
regressor.fit(X_train,y_train)
```

```python
regressor.intercept_
```

```python
regressor.n_support_
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
import pickle
filename="finalized_model_svm.sav"
pickle.dump(regressor,open(filename,'wb'))
```

```python
#provide all 5 inputs as we trained with a model that had 5 input variables
loaded_model=pickle.load(open("finalized_model_svm.sav",'rb'))
result=loaded_model.predict([[1234,345,4565,1,0]])
```
