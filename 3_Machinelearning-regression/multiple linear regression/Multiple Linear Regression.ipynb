```python
#importing library
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
#converting categorical column -nominal value with 1 hot encoding with data expansion, get dummies function will convert categorical to  numerical data 
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
#getting column names 
dataset.columns
```

```python
#displaying indenpendent or input variables
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
independent
```

```python
#displaying output or dependent variables
dependent=dataset[["Profit"]]
```

```python
#splitting training and test data for input and output
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
#model creation using training data
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
LinearRegression()
```

```python
#checking model's weight
weight=regressor.coef_
weight
```

```python
#checking origin (without any profit in business)
bais=regressor.intercept_
bais
```

```python
y_pred=regressor.predict(X_test)
```

```python
#evaluating rscore
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
import pickle
filename="finalized_model_Mul_linear.sav"
pickle.dump(regressor,open(filename,'wb'))
```
